```python
def print_board(board):
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")

def check_winner(board, player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Horizontal
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Vertical
        [0, 4, 8], [2, 4, 6]              # Diagonal
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

def is_board_full(board):
    return all(cell in ['X', 'O'] for cell in board)

def tic_tac_toe():
    board = [' '] * 9
    current_player = 'X'
    
    while True:
        print_board(board)
        
        # Get player input
        try:
            move = int(input(f"Player {current_player}, enter your move (1-9): ")) - 1
            if move < 0 or move > 8 or board[move] != ' ':
                print("Invalid move. Try again.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 9.")
            continue

        # Make the move
        board[move] = current_player

        # Check for a winner
        if check_winner(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break

        # Check for a draw
        if is_board_full(board):
            print_board(board)
            print("It's a draw!")
            break

        # Switch players
        current_player = 'O' if current_player == 'X' else 'X'

# Run the game
tic_tac_toe()
```

```python
import math

def minmax(current_index, node_index, scores, max_turn, target_depth):
    # If we reach the target depth, return the score at the node index
    if current_index == target_depth:
        return scores[node_index]
    
    # If it's the maximizer's turn
    if max_turn:
        return max(minmax(current_index + 1, node_index * 2, scores, False, target_depth),
                   minmax(current_index + 1, node_index * 2 + 1, scores, False, target_depth))
    # If it's the minimizer's turn
    else:
        return min(minmax(current_index + 1, node_index * 2, scores, True, target_depth),
                   minmax(current_index + 1, node_index * 2 + 1, scores, True, target_depth))

# Scores array representing the leaves of the game tree
scores = [3, 5, 2, 9, 12, 5, 23, 23]

# Calculate the tree depth
tree_depth = int(math.log(len(scores), 2))

# Print the optimal value that the maximizer can obtain
print(minmax(0, 0, scores, True, tree_depth))
```

```python
from itertools import permutations

def solve_cryptarithmetic():
    # Unique letters in the puzzle
    letters = 'SENDMORY'

    # All permutations of digits (0-9) for 8 unique letters
    for perm in permutations('0123456789', len(letters)):
        # Mapping letters to digits
        s, e, n, d, m, o, r, y = perm

        # Check if the mapping is valid (no leading zeroes for SEND, MORE, and MONEY)
        if s == '0' or m == '0':
            continue

        # Convert the letters into numbers
        send = int(s + e + n + d)
        more = int(m + o + r + e)
        money = int(m + o + n + e + y)

        # Check if the equation holds
        if send + more == money:
            print(f"SEND = {send}")
            print(f"MORE = {more}")
            print(f"MONEY = {money}")
            return

solve_cryptarithmetic()
```

```python
import torch
from transformers import VisionEncoderDecoderModel, AutoTokenizer
from sentence_transformers import SentenceTransformer, util
import pytesseract
import cv2
import re
from PIL import Image
import numpy as np

# Load Donut OCR model
donut_model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-base")
donut_tokenizer = AutoTokenizer.from_pretrained("naver-clova-ix/donut-base")

# Load Sentence-BERT model for recipe matching
sbert_model = SentenceTransformer("all-MiniLM-L6-v2")

# Set Tesseract OCR path (Update if needed)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Recipe dataset (Example)
recipe_list = [
    "Tomato Pasta with Cheese",
    "Spicy Chicken Curry",
    "Vegetable Stir Fry",
    "Grilled Fish with Lemon",
]

#  Function to preprocess receipt image
def preprocess_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
    image = cv2.GaussianBlur(image, (5, 5), 0)  # Apply Gaussian blur
    return Image.fromarray(image)  # Convert back to PIL image

#  Function to extract text using Donut
def extract_text_donut(image_path):
    image = preprocess_image(image_path)
    pixel_values = donut_model.image_processor(image, return_tensors="pt").pixel_values
    outputs = donut_model.generate(pixel_values)
    extracted_text = donut_tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    return extracted_text

#  Function to extract text using Tesseract OCR
def extract_text_tesseract(image_path):
    image = preprocess_image(image_path)
    extracted_text = pytesseract.image_to_string(image)
    return extracted_text

#  Function to extract expiry date from text using regex
def extract_expiry_date(text):
    date_pattern = r"\b(\d{1,2}/\d{1,2}/\d{4}|\d{4}-\d{2}-\d{2})\b"  # Matches DD/MM/YYYY or YYYY-MM-DD
    match = re.search(date_pattern, text)
    return match.group(0) if match else "Not Found"

#  Function to match ingredients to recipes using Sentence-BERT
def get_recipe_sbert(ingredients, recipe_list):
    ingredient_embedding = sbert_model.encode(ingredients, convert_to_tensor=True)
    best_recipe, best_score = None, 0

    for recipe in recipe_list:
        recipe_embedding = sbert_model.encode(recipe, convert_to_tensor=True)
        score = util.pytorch_cos_sim(ingredient_embedding, recipe_embedding).item()
        if score > best_score:
            best_recipe, best_score = recipe, score

    return best_recipe, best_score

#  Function to match ingredients using simple keyword matching
def get_recipe_rule_based(ingredients, recipe_list):
    best_match, max_overlap = None, 0

    for recipe in recipe_list:
        common_words = set(ingredients.split()) & set(recipe.split())
        if len(common_words) > max_overlap:
            best_match, max_overlap = recipe, len(common_words)

    return best_match

#  Main Execution
def main(image_path):
    print("\n Processing Image:", image_path)

    # Extract text using both methods
    text_donut = extract_text_donut(image_path)
    text_tesseract = extract_text_tesseract(image_path)

    # Extract expiry dates
    expiry_donut = extract_expiry_date(text_donut)
    expiry_tesseract = extract_expiry_date(text_tesseract)

    # Extract ingredients list
    ingredients_donut = " ".join(re.findall(r'\b\w+\b', text_donut))
    ingredients_tesseract = " ".join(re.findall(r'\b\w+\b', text_tesseract))

    # Get recipe recommendations
    best_recipe_sbert, sbert_score = get_recipe_sbert(ingredients_donut, recipe_list)
    best_recipe_rule_based = get_recipe_rule_based(ingredients_tesseract, recipe_list)

    # Print Results
    print("\n OCR Results:")
    print("Donut Extracted Text:\n", text_donut)
    print("Tesseract Extracted Text:\n", text_tesseract)

    print("\n Expiry Date Extraction:")
    print("Donut:", expiry_donut)
    print("Tesseract:", expiry_tesseract)

    print("\n Recipe Matching:")
    print(f"Donut + SBERT Best Recipe: {best_recipe_sbert} (Similarity Score: {sbert_score:.2f})")
    print(f"Tesseract + Rule-Based Best Recipe: {best_recipe_rule_based}")

# Run the program (Replace 'receipt_sample.jpg' with your image)
main("receipt_sample.jpg")
```
